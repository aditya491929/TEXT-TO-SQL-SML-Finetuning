```python
import json, os, tarfile
from pathlib import Path

SPIDER_DIR = Path("/path/to/local/spider")  # has dev.json, tables.json, database/
OUT_TAR = Path("spider_dev_dbs.tar.gz")

dev = json.load(open(SPIDER_DIR/"dev.json"))
need = sorted({ex["db_id"] for ex in dev})

db_root = SPIDER_DIR/"database"
missing = [d for d in need if not (db_root/d/f"{d}.sqlite").exists()]
print("Missing:", missing)

with tarfile.open(OUT_TAR, "w:gz") as tar:
    for d in need:
        tar.add((db_root/d), arcname=f"database/{d}")

print("Wrote", OUT_TAR, "with", len(need), "DB folders")
```
